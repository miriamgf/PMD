```python
def find_sepsis():
  #UNIDADES
  list_sepsis=[]
  list_not_sepsis=[]
  contador = 0
  for i in range(1,10):
      name="/content/drive/My Drive/Colab Notebooks/training_setA/p00000%d.psv" %i
      df= pd.read_csv(name, sep = "|") 
      y=list(df.iloc[:,-1])
      if 1 in y:
            if name not in list_sepsis:
                print(name,'Hay sepsis')
                list_sepsis.append(name)
      else:
          print(name,'No hay sepsis')
          list_not_sepsis.append(name)
  #DECENAS
  for i in range(10,100):
      name="/content/drive/My Drive/Colab Notebooks/training_setA/p0000%d.psv" %i
      df= pd.read_csv(name, sep = "|") 
      y=list(df.iloc[:,-1])
      if 1 in y:
            if name not in list_sepsis:
                print(name,'Hay sepsis')
                list_sepsis.append(name)
      else:
          print(name,'No hay sepsis')
          list_not_sepsis.append(name)

  #CENTENAS
  for i in range(100,500):
      name="/content/drive/My Drive/Colab Notebooks/training_setA/p000%d.psv" %i
      df= pd.read_csv(name, sep = "|") 
      y=list(df.iloc[:,-1])
      if 1 in y:
            if name not in list_sepsis:
                print(name,'Hay sepsis')
                list_sepsis.append(name)
      else:
          print(name,'No hay sepsis')
          list_not_sepsis.append(name)

  return(list_sepsis, list_not_sepsis)


# Funcion para ver que positivos estan cualificados: mas de 18 ceros antes de sepsis
def find_sepsis_over18zeros(list_sepsis):
    list_sepsis_over18=[]
    for i in list_sepsis:
        df = pd.read_csv(i, sep = "|")
        y=list(df.iloc[:,-1])
        contador=0
        for elemento in y:
            if elemento == 0:
                contador+=1
                if contador >= 18:
                    if i not in list_sepsis_over18:
                        list_sepsis_over18.append(i)
    return(list_sepsis_over18)  

# Funcion para interpolar los datos
def interpolate(df):
  #Declarar todas las columnas
  #VITAL SIGNS
  HR=df['HR']
  O2Sat=df['O2Sat']
  Temp=df['Temp']
  SBP=df['SBP']
  MAP=df['MAP']
  DBP=df['DBP']
  Resp=df['Resp']
  EtCO2=df['EtCO2']
  #LABORATORY VALUES
  BaseExcess=df['BaseExcess']
  HCO3=df['HCO3']
  FiO2=df['FiO2']
  pH=df['pH']
  PaCO2=df['PaCO2']
  SaO2=df['SaO2']
  AST=df['AST']
  BUN=df['BUN']
  Alkalinephos=df['Alkalinephos']
  Calcium=df['Calcium']
  Chloride=df['Chloride']
  Creatinine=df['Creatinine']
  Bilirubin_direct=df['Bilirubin_direct']
  Glucose=df['Glucose']
  Lactate=df['Lactate']
  Magnesium=df['Magnesium']
  Phosphate=df['Phosphate']
  Potassium=df['Potassium']
  Bilirubin_total=df['Bilirubin_total']
  TroponinI=df['TroponinI']
  Hct=df['Hct']
  Hgb=df['Hgb']
  PTT=df['PTT']
  WBC=df['WBC']
  Fibrinogen=df['Fibrinogen']
  Platelets=df['Platelets']
  #DEMOGRAPHICS
  Age=df['Age']
  Gender=df['Gender']
  Unit1=df['Unit1']
  Unit2=df['Unit2']
  HospAdmTime=df['HospAdmTime']
  ICULOS=df['ICULOS']
  #Outcome
  SepsisLabel=df['SepsisLabel']

  #Guardamos nombres de las columnas divididos en tres tipos de variables: signos vitales, valores de lab y datos demográficos
  #VITAL SIGNS
  HR2 = HR.fillna(method='backfill')
  HR3 = HR2.fillna(method='ffill')
  O2Sat2 = O2Sat.fillna(method='backfill')
  O2Sat3 = O2Sat2.fillna(method='ffill')
  Temp2 = Temp.fillna(method='backfill')
  Temp3 = Temp2.fillna(method='ffill')      
  SBP2 = SBP.fillna(method='backfill')
  SBP3 = SBP2.fillna(method='ffill')
  MAP2 = MAP.fillna(method='backfill')
  MAP3 = MAP2.fillna(method='ffill')
  DBP2 = DBP.fillna(method='backfill')
  DBP3 = DBP2.fillna(method='ffill')
  Resp2 = Resp.fillna(method='backfill')
  Resp3 = Resp2.fillna(method='ffill')
  EtCO2_2 = EtCO2.fillna(method='backfill')
  EtCO2_3 = EtCO2_2.fillna(method='ffill')
  #LABORATORY VALUES
  BaseExcess2 = BaseExcess.interpolate(method='linear')
  BaseExcess3 = BaseExcess2.interpolate(method='linear',limit_direction='backward')
  HCO3_2 = HCO3.interpolate(method='linear')
  HCO3_3 = HCO3_2.interpolate(method='linear',limit_direction='backward')
  FiO2_2 = FiO2.interpolate(method='linear')
  FiO2_3 = FiO2_2.interpolate(method='linear',limit_direction='backward')
  pH2 = pH.interpolate(method='linear')
  pH3 = pH2.interpolate(method='linear',limit_direction='backward')
  PaCO2_2 = PaCO2.interpolate(method='linear')
  PaCO2_3 = PaCO2_2.interpolate(method='linear',limit_direction='backward')
  SaO2_2 = SaO2.interpolate(method='linear')
  SaO2_3 = SaO2_2.interpolate(method='linear',limit_direction='backward')
  AST2 = AST.interpolate(method='linear')
  AST3 = AST2.interpolate(method='linear',limit_direction='backward')
  BUN2 = BUN.interpolate(method='linear')
  BUN3 = BUN2.interpolate(method='linear',limit_direction='backward')
  Alkalinephos2 = Alkalinephos.interpolate(method='linear')
  Alkalinephos3 = Alkalinephos2.interpolate(method='linear',limit_direction='backward')
  Calcium2 = Calcium.interpolate(method='linear')
  Calcium3 = Calcium2.interpolate(method='linear',limit_direction='backward')
  Chloride2 = Chloride.interpolate(method='linear')
  Chloride3 = Chloride2.interpolate(method='linear',limit_direction='backward')
  Creatinine2 = Creatinine.interpolate(method='linear')
  Creatinine3 = Creatinine2.interpolate(method='linear',limit_direction='backward')
  Bilirubin_direct2 = Bilirubin_direct.interpolate(method='linear')
  Bilirubin_direct3 = Bilirubin_direct2.interpolate(method='linear',limit_direction='backward')
  Glucose2 = Glucose.interpolate(method='linear')
  Glucose3 = Glucose2.interpolate(method='linear',limit_direction='backward')
  Lactate2 = Lactate.interpolate(method='linear')
  Lactate3 = Lactate2.interpolate(method='linear',limit_direction='backward')
  Magnesium2 = Magnesium.interpolate(method='linear')
  Magnesium3 = Magnesium2.interpolate(method='linear',limit_direction='backward')
  Phosphate2 = Phosphate.interpolate(method='linear')
  Phosphate3 = Phosphate2.interpolate(method='linear',limit_direction='backward')
  Potassium2 = Potassium.interpolate(method='linear')
  Potassium3 = Potassium2.interpolate(method='linear',limit_direction='backward')
  Bilirubin_total2 = Bilirubin_total.interpolate(method='linear')
  Bilirubin_total3 = Bilirubin_total2.interpolate(method='linear',limit_direction='backward')
  TroponinI2 = TroponinI.interpolate(method='linear')
  TroponinI3 = TroponinI2.interpolate(method='linear',limit_direction='backward')
  Hct2 = Hct.interpolate(method='linear')
  Hct3 = Hct2.interpolate(method='linear',limit_direction='backward')
  Hgb2 = Hgb.interpolate(method='linear')
  Hgb3 = Hgb2.interpolate(method='linear',limit_direction='backward')
  PTT2 = PTT.interpolate(method='linear')
  PTT3 = PTT2.interpolate(method='linear',limit_direction='backward')
  WBC2 = WBC.interpolate(method='linear')
  WBC3 = WBC2.interpolate(method='linear',limit_direction='backward')
  Fibrinogen2 = Fibrinogen.interpolate(method='linear')
  Fibrinogen3 = Fibrinogen2.interpolate(method='linear',limit_direction='backward')
  Platelets2 = Platelets.interpolate(method='linear')
  Platelets3 = Platelets2.interpolate(method='linear',limit_direction='backward')

# Recopilar todas las variables una vez interpoladas
  matrix = pd.DataFrame({"HR" : HR3, "O2Sat": O2Sat3, "Temp": Temp3, "SBP":SBP3, "MAP":MAP3, "DBP":DBP3, "Resp":Resp3, "EtCO2":EtCO2_3, "BaseExcess":BaseExcess3, "HCO3":HCO3_3, "FiO2":FiO2_3, "pH":pH3, "PaCO2":PaCO2_3, "SaO2":SaO2_3, 
                      "AST":AST3, "BUN":BUN3, "Alkalinephos":Alkalinephos3, "Calcium":Calcium3, "Chloride":Chloride3, "Creatinine":Creatinine3, "Bilirubin_direct":Bilirubin_direct3, "Glucose":Glucose3, "Lactate":Lactate3, 
                      "Magnesium":Magnesium3, "Phosphate":Phosphate3, "Potassium":Potassium3, "Bilirubin_total":Bilirubin_total3, "TroponinI":TroponinI3, "Hct":Hct3, "Hgb":Hgb3, "PTT":PTT3, "WBC":WBC3, "Fibrinogen":Fibrinogen3,
                      "Platelets":Platelets3, "Age": Age, "Gender":Gender, "Unit1":Unit1, "Unit2":Unit2, "HospAdmTime":HospAdmTime, "ICULOS":ICULOS, "SepsisLabel":SepsisLabel},columns=["HR", "O2Sat", "Temp", "SBP", "MAP","DBP", "Resp", "EtCO2", "BaseExcess", "HCO3", "FiO2", "pH", "PaCO2", "SaO2", 
                      "AST", "BUN", "Alkalinephos", "Calcium", "Chloride", "Creatinine", "Bilirubin_direct", "Glucose", "Lactate", 
                      "Magnesium", "Phosphate", "Potassium", "Bilirubin_total", "TroponinI", "Hct", "Hgb", "PTT", "WBC", "Fibrinogen",
                      "Platelets", "Age", "Gender", "Unit1", "Unit2", "HospAdmTime", "ICULOS", "SepsisLabel"])
  
  return matrix
  # Funcion para seleccionar indices de positivos (y 6 horas antes) y negativos
def seleccion_indices(matrix):
  non_sepsis=[]
  positive_sepsis=[]
  negative_sepsis=[]
  
  sepsis_col=list(matrix["SepsisLabel"])
  for position_sepsis in range(len(sepsis_col)):
    if sepsis_col[position_sepsis]==1:  #Position_sepsis: cuando comienza la sepsis
      break

  for j in range(position_sepsis-6,position_sepsis+6):
     positive_sepsis.append(j) #Tanto 6h hacia atras como 6h hacia delante desde que empieza la sepsis

  for k in range(0,position_sepsis-6):
     non_sepsis.append(k) #Todos los negativos hasta 6 horas antes de la sepsis

  while len(negative_sepsis)<12 :
    random=np.random.choice(non_sepsis) #Seleccionar 12 indices aleatorios con sepsis negativo
    if random not in negative_sepsis:
        negative_sepsis.append(random)

  return(negative_sepsis, positive_sepsis)

def create_outcome(matrix, positive_sepsis, negative_sepsis):
  matrix["Outcome"]= "NaN" #Rellenar entera con Nan

  for v in positive_sepsis:
    matrix.iloc[v,-1]=1

  for w in negative_sepsis:
    matrix.iloc[w,-1]=0
  
  finalmatrix = matrix[matrix["Outcome"]!="NaN"]
  return(finalmatrix)
```

```python
variables = ["HR", "O2Sat", "Temp", "SBP", "MAP","DBP", "Resp", "EtCO2", "BaseExcess", "HCO3", "FiO2", "pH", "PaCO2", "SaO2", 
                      "AST", "BUN", "Alkalinephos", "Calcium", "Chloride", "Creatinine", "Bilirubin_direct", "Glucose", "Lactate", 
                      "Magnesium", "Phosphate", "Potassium", "Bilirubin_total", "TroponinI", "Hct", "Hgb", "PTT", "WBC", "Fibrinogen",
                      "Platelets", "Age", "Gender", "Unit1", "Unit2", "HospAdmTime", "ICULOS", "SepsisLabel", "Outcome", "IdPaciente"]

empty = pd.DataFrame(columns=variables)

for i in range(24): #Numero de pacientes 
    df = pd.read_csv(index_sepsis[i],sep='|') #Abrimos estos archivos
    matrix=interpolate(df) #Interpolamos
    matrix2=matrix.fillna(0) #Los NaN que no se han podido interpolar los rellenamos con 0
    matrix2["IdPaciente"]=i #Creamos nuevo identificador de paciente
    indexes = seleccion_indices(matrix2) #
    negative_sepsis = indexes[0] #Pacientes negativos
    positive_sepsis = indexes[1] #Pacientes positivos
    finalmatrix = create_outcome(matrix2, positive_sepsis, negative_sepsis) #Creamos nueva variable Outcome con lo que consideramos positivo y negativo
    empty=pd.concat([empty, finalmatrix], sort=False) #Concatenamos los dataframes de los pacientes secuencialmente   
```
