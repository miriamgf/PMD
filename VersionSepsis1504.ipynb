```python
def find_sepsis_over18zeros(list_sepsis):
    list_sepsis_over18=[]
    for i in list_sepsis:
        df = pd.read_csv(i, sep = "|")
        y=list(df.iloc[:,-1])
        contador=0
        for elemento in y:
            if elemento == 0:
                contador+=1
                if contador >= 18:
                    if i not in list_sepsis_over18:
                        list_sepsis_over18.append(i)
    return(list_sepsis_over18)
        
```

```python
def seleccion_indices(matrix):
  # Seleccionar indices de positivos (y 6 horas antes) y negativos
  sepsis_col=list(matrix["SepsisLabel"])
  for position_sepsis in range(len(sepsis_col)):
    if sepsis_col[position_sepsis]==1:
      break

  non_sepsis=[]
  positive_sepsis=[]
  negative_sepsis=[]

  for j in range(position_sepsis-6,position_sepsis+6):
     positive_sepsis.append(j)

  for k in range(0,position_sepsis-6):
     non_sepsis.append(k)

  # Seleccionar 12 indices aleatorios con sepsis negativo
  while len(negative_sepsis)<12 :
    random=np.random.choice(non_sepsis)
    if random not in negative_sepsis:
        negative_sepsis.append(random)

  return(negative_sepsis, positive_sepsis)

```

```python
def create_outcome(matrix, positive_sepsis, negative_sepsis):
  # Hacer nueva columna con nuevos resultados de sepsis
  matrix["Outcome"]= "NaN"

  for v in positive_sepsis:
    matrix.iloc[v,-1]=1

  for w in negative_sepsis:
    matrix.iloc[w,-1]=0
  
  finalmatrix = matrix[matrix["Outcome"]!="NaN"]
  return(finalmatrix)
```
